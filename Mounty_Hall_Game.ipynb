```
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import random
```

```
def prob_monty(n, switch):
    '''
    Runs monty_hall game n times
    Returns dictionary with number of wins and loses
    '''
    result = {"win":0, "lose":0}
    for each in range(n):
        result[mounty_hall(switch)] += 1

    return result

def mounty_hall(switch):
    '''
    switch option - True or False
    Mounty hall (in)famous television game show in the form of a probability
    puzzle.
    Program runs the game once and returns win or lose
    '''
    doors = [True, False, False] # creating three doors
    my_door = random.choice(doors) # randomly pick a door
    doors.remove(my_door) # remove choosen door from list of doors
    doors.remove(False) # remove a false door

    if switch:
        my_door = doors[0]

    if my_door:
        return "win"
    else:
        return "lose"
```

```
# Run 100 experiments of 1000 games each switiching doors
switch = []
for each in range(100):
  switch.append(prob_monty(1000, True))
df_switch = pd.DataFrame(switch)

# Run 100 experiments of 1000 games each not switiching doors
stay = []
for each in range(100):
  stay.append(prob_monty(1000, False))
df_stay = pd.DataFrame(stay)
```

```
fig, ((ax1, ax2)) = plt.subplots(nrows=1, ncols=2, figsize=(15,5))
sns.histplot(data=df_switch, x='win', kde=True, ax=ax1)
ax1.set_title('Trocando as portas')
sns.histplot(data=df_stay, x='win', kde=True, color='r', ax=ax2)
ax1.set_title('Não trocando as portas')
```
